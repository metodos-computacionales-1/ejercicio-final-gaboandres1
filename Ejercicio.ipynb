```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.system('g++ Pendulum.cpp -o Pendulum.x')
os.system('./Pendulum.x 10000 1 4 0 0 0 > data.dat')
```

```python
data = np.loadtxt('data.dat')
plt.plot(data[:,0], data[:,1], label='$\Theta$')
plt.plot(data[:,0], data[:,2], label='$\omega$')
plt.xlabel('Time')
plt.legend()
plt.show()
```

```python
F = [1.4, 1.44, 1.465]
os.system('./Pendulum.x 5000 1 4 0 0 0 > data1.dat')
os.system('./Pendulum.x 5000 1 4 4 0 0 > data2.dat')
os.system('./Pendulum.x 5000 1 4 6 5 0 > data3.dat')
data1 = np.loadtxt('data1.dat')
data2 = np.loadtxt('data2.dat')
data3 = np.loadtxt('data3.dat')

plt.plot(data1[:,1], data1[:,2], label='$F_D=1.4$')
plt.plot(data2[:,1], data2[:,2], label='$F_D=1.44$')
plt.plot(data3[:,1], data3[:,2], label='$F_D=1.465$')
plt.xlabel('$\Theta$')
plt.ylabel('$\omega$')
plt.legend()
plt.show()
```

```python
FD = np.arange(1.35, 1.5, 0.0001)
Theta = []
for Fd in FD:
    if(len(str(Fd)) == 3):
        os.system('./Pendulum.x 50000 '+str(Fd)[0]+' '+str(Fd)[2]+' 0 0 0 > data.dat')
    elif(len(str(Fd)) == 4):
        os.system('./Pendulum.x 50000 '+str(Fd)[0]+' '+str(Fd)[2]+' '+str(Fd)[3]+' 0 0 > data.dat')
    elif(len(str(Fd)) == 5):
        os.system('./Pendulum.x 50000 '+str(Fd)[0]+' '+str(Fd)[2]+' '+str(Fd)[3]+' '+str(Fd)[4]+' 0 > data.dat')
    else:
        os.system('./Pendulum.x 50000 '+str(Fd)[0]+' '+str(Fd)[2]+' '+str(Fd)[3]+' '+str(Fd)[4]+' '+str(Fd)[5]+' > data.dat')
    data = np.loadtxt('data.dat')
    Theta.append(data[-1,2])    
plt.scatter(FD, Theta)
plt.xlabel('$F_D$')
plt.ylabel('$\Theta$')
plt.plot()
```
